```python

import collections

#longest subarray
def maxLength(a, k):
    s, c, mx = 0,0,0
    for i in range(len(a)):
        if s+a[i]<=k:
            s += a[i]
            c+=1
        else:
            if s != 0:
                s = s - a[i-c] + a[i]
        mx = max(c, mx)
    return mx
print("longest subarray", maxLength([1,2,3],4))
#Visiting cities
def minimumCost(red, blue, blueCost):
    red_l = len(red)
    res = []
    dp = [[0]*2]*(red_l+1)
    dp[0][0] = 0
    dp[0][1] = blueCost
    res.append(0)
    for i in range(red_l):
        dp[i][0] = min(dp[i-1][0]+red[i-1], dp[i-1][1]+red[i-1])
        dp[i][1] = min(dp[i-1][1]+blue[i-1], dp[i-1][0]+blue[i-1]+blueCost)
        res.append(min(dp[i][0],dp[i][1]))
    return res
    
print("visiting cites:",minimumCost([5],[3],1))
#shrinking number line
def smallestRangeII(nums, k):
    nums.sort()
    left, right = nums[0],nums[-1]
    res = right - left
    for i in range(len(nums)-1):
        maxi = max(nums[i]+k, right-k)
        mini = min(left+k, nums[i+1]-k)
        res = min(res, maxi-mini)
    return res

#K-subarrays
def subarraysDivByK(nums, k):
    count = [1]+[0]*k
    rem = 0
    res = 0
    for a in nums:
        rem = (rem+a)%k
        res += count[rem]
        count[rem] += 1
    return res

#Sprint Training
def mostVisited(sprints, n):
    temp = [0]*(n+1)
    score = 0
    for i in range(1, len(sprints)):
        start = min(sprints[i], sprints[i-1])
        end = max(sprints[i], sprints[i-1])
        temp[start]+=1
        if end < n:
            temp[end+1] -= 1
    for i in range(1, n+1):
        temp[i] += temp[i-1]
        if temp[i] > score:
            score = temp[i]
            res = i
 
    return res

# do they belong 
import math 

def area(x1, y1, x2, y2, x3, y3):

    return abs((x1*(y2-y3) + x2*(y3-y1)+ x3*(y1-y2))/2.0)

def isInside(x1, y1, x2, y2, x3,  y3, x, y): 

    A = area (x1, y1, x2, y2, x3, y3); 

    A1 = area (x, y, x2, y2, x3, y3); 

    A2 = area (x1, y1, x, y, x3, y3); 

    A3 = area (x1, y1, x2, y2, x, y); 
	
    return (A == A1 + A2 + A3); 

def distance(x1,  y1,  x2,  y2):
    return ( math.sqrt( pow((x2 - x1), 2) + pow((y2 - y1), 2) ) )

def isValid(x1, y1, x2, y2, x3, y3):
    a = distance(x1, y1, x2, y2)
    b = distance(x2, y2, x3, y3) 
    c = distance(x3, y3, x1, y1)
    h = False 
    if ((a+b)>c) and ((a+c)>b) and ((c+b)>a):
        h = True
    return h 

#get the groups
class UF:
    def __init__(self, n) -> None:
        self.root = list(range(n+1))
        self.count = [0]+[1]*n

    def find(self, x):
        if self.root[x] != x: 
            self.root[x] = self.find(self.root[x])
        return self.root[x]
    
    def union(self, n1, n2):
        r1 = self.find(n1)
        r2 = self.find(n2)
        if r1 != r2:
            if self.count[n1] > self.count[n2]:
                self.root[r2] = r1
                self.count[r1] += self.count[r2]
            else:
                self.root[r1] = r2
                self.count[r2] += self.count[r1]

def findGroupSize(n, queryType, student1, student2):
    uf = UF(n)
    for i in range(len(student1)):
        if queryType[i] == 'Friend':
            uf.union(student1[i], student2[i])
        elif queryType[i] == 'Total':
            if uf.find(student1[i]) == uf.find(student2[i]):
                return uf.count[student1[i]]
            else:
                return uf.count[student1[i]]+uf.count[student2[i]]

print("get the groups: ",findGroupSize(3,['Friend','Total'],[1,2],[2,3]))


```
